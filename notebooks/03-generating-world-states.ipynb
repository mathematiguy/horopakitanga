# Generating world states

In this notebook we'll be building an environment containing an arrangement of coloured sticks of different lengths. The environment is analogous to the use of 'cuisennaire rods' in the Silent Way, a language teaching methodology by Caleb Gategno, which was adapted to the Māori language by Dame Kāterina Te Heikōkō Mataira and Ngoingoi Pēwhairangi.

The goal will be to generate a large number of these world states and then I can write down a bunch of utterances for them to build a training set for fine-tuning a model.
